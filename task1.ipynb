Imports

```python
import random
import matplotlib.pyplot as plt
```

Generating heights

```python
heights = [random.randint(0, 26) for _ in range(26)]
print("Heights:", heights)
```

Finding the deepest "lake"

```python
def deepest_lake(heights):
    n = len(heights)
    max_depth = 0
    deepest_i = deepest_j = 0

    for i in range(n):
        for j in range(i+1, n):
            waterlevel = min(heights[i], heights[j])
            for k in range(i+1, j):
                depth = waterlevel - heights[k]
                if depth > max_depth:
                    max_depth = depth
                    deepest_i, deepest_j = i, j

    return max_depth, deepest_i, deepest_j
```

Function plot_deepest_lake

```python
def plot_deepest_lake(heights, deepest_i, deepest_j):
    n = len(heights)
    x = list(range(n))

    plt.figure(figsize=(10, 10))
    plt.plot (x, heights, color='black')

    if deepest_i is not None and deepest_j is not None:
        lake_x = list(range(deepest_i, deepest_j +1))
        lake_y = heights[deepest_i:deepest_j+1]
        plt.plot(lake_x, lake_y, color='blue')
        plt.fill(lake_x, lake_y, [0] * len(lake_x), color='blue', alpha=0.2)

    plt.ylabel('Height')
    plt.xlabel('Widgth')
    plt.grid(True)
    plt.show()
```

Printing the results

```python
max_depth, deepest_i, deepest_j = deepest_lake(heights)
print ("Depth of the lake is: ", max_depth)
```

Visualization

```python
plot_deepest_lake(heights, deepest_i, deepest_j)
```
