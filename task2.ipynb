Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generation of matrix

```python
np.random.seed(42)

grid = np.random.randint(0, 2, size=(26, 26))

print(grid)
```

Function of counting "alive" neighbors

```python
def neighborhood(grid, row, col):

    neighbors = 0

    for i in range(row-1, row+2):
        for j in range(col-1, col+2):
            if i == row and j == col:
                continue

            if 0 <= i < grid.shape[0] and 0 <= j < grid.shape[1]:
                neighbors += grid[i, j]

    return neighbors
```

Function of one step

```python
def step(grid):

    new_grid = grid.copy()
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            alive = neighborhood(grid, i, j)

            if grid[i, j] == 1:
                if alive < 2 or alive > 3:
                    new_grid[i, j] = 0

            else:
                if alive == 3:
                    new_grid[i, j] = 1

    return new_grid
```

Start of 20 iterations

```python
steps = 20

for _ in range(steps):
    grid = step(grid)

print(grid)
```

Visualization

```python
def visualize_grid(grid, title=""):
    grid = np.asarray(grid).astype(int)

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.imshow(grid, cmap="Greens")

    ax.set_xticks(np.arange(-0.5, 26, 1))
    ax.set_yticks(np.arange(-0.5, 26, 1))
    ax.grid(True, color="gray", linewidth=0.5)

    ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
    ax.set_title(title)

    plt.show()
```

Final state, after 20 steps

```python
visualize_grid(grid, title="State after 20 steps")
```
