Given data

```python
coin_probs = [0.12, 0.27, 0.21, 0.96]
priors = [0.25, 0.25, 0.25, 0.25]
observations = ['H', 'H', 'H', 'T', 'H', 'T', 'H', 'H', 'H']
results = []
```

Iterating through each flip

```python
for flip in observations:
    new_priors = []
    total_probability = 0

    for i in range(4):
        if flip == 'H':
            likelihood = coin_probs[i]
        else:
            likelihood = 1 - coin_probs[i]

        val = likelihood * priors[i]
        new_priors.append(val)
        total_probability += val

    for i in range(4):
        priors[i] = new_priors[i] / total_probability

    next_flip = 0
    for i in range(4):
        next_flip += priors[i] * coin_probs[i]

    results.append(round(next_flip, 2))
```

Results

```python
print(results)
```
